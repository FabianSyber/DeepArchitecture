```python
import cv2
import os
import glob
import numpy as np
import PIL.Image
import re
import pathlib
import shutil

print('running')

data_dir = pathlib.Path('E:/DataSets/cubicasa5kv1.1')
data_dir_str = 'E:/DataSets/cubicasa5kv1.1'

for file in data_dir.glob('*/function_map_new.png'):
    file = str(file)
    file = file.replace('\\', '/')
    res = re.findall("(\w+)/function_map_new.png", file)
    filename = res[0]
    
    img = os.path.join(data_dir, filename, 'function_map_new.png')

    img = cv2.imread(img, 1)

    # Read the image
    height, width, channels = img.shape

    # convert to hsv
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    gwash = ~img
    gwashBW = cv2.cvtColor(gwash, cv2.COLOR_BGR2GRAY) #change to grayscale

    ret, thresh = cv2.threshold(gwashBW,200,255,cv2.THRESH_BINARY)
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) #find contours with simple approximation
    
    # We remove any floorplans that consists of more than one structure
    if(len(contours)>2):
        print('\nthis one should be removed ' + filename)
    else:
        # Get masks of colors that we want to sort on
        lower = np.array([0,0,175])
        upper = np.array([50,50,190])
        mask_grey = cv2.inRange(hsv, lower, upper)

        lower = np.array([0,0,0])
        upper = np.array([10,10,10])
        mask_black = cv2.inRange(hsv, lower, upper)

        pixels = height * width

        # Get counts of black and grey pixels so that we can check percentages and mark those that are interesting
        count_black = np.count_nonzero(mask_black)
        count_grey = np.count_nonzero(mask_grey)

        # these are the amount of pixels that are colored and not background
        plan_pixels = pixels - count_black
        
        # If 25% of image is grey(unknown) we remove it from this dataset
        if((count_grey/plan_pixels)>0.25):
            print('\nthis one should be removed ' + filename)
        else:
            print('\nthis one should be kept ' + filename + ' moving files to cubicasa5kv1.2')
    
            os.mkdir('E:/DataSets/cubicasa5kv1.2/' + filename)
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/clean.png', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/clean.png')
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/F1_original.png', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/F1_original.png')
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/function_map_new.png', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/function_map_new.png')
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/mask.png', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/mask.png')
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/outline.png', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/outline.png')
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/model.svg', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/model.svg')
            shutil.copy('E:/DataSets/cubicasa5kv1.1/' + filename + '/function_map_no_icon.png', 'E:/DataSets/cubicasa5kv1.2/' + filename + '/function_map_no_icon.png')
            
print('done')
```
